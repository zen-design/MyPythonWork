#### PrintAllPairs(n)
    
    輸入 : 整數 n
    效果 : 印出所有(i, j)的pair, 其中 1 <= i,j <=n

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
def PrintAllPairs(n):
    pairs = []
    for i in range(1, n + 1):
        for j in range(1, n + 1):
            tuple = (i, j)
            pairs.append(tuple)
            
test_num_list = []
used_time = [0]

for i in range(1, 11):
    test_num = 100*i
    test_num_list.append(test_num)

for j in range(0, 10):
    start_time = time.time()
    PrintAllPairs(test_num_list[j])
    end_time = time.time()
    used_time.append(end_time - start_time)

plt.plot(used_time)
plt.xlabel('n')
plt.ylabel('used_time')
#print出來檔案太大無法上傳    
```
